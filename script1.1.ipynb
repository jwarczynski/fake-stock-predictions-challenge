$$
\begin{align*}
\text{ECM model:} \\
\text{min } f_{\text{risk}} &= \sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \sigma_{ij}, \\
\sum_{i=1}^{n} w_i r_i &\geq \text{return}, \\
\sum_{i=1}^{n} w_i &= 1, \\
w_i &\in [0,1] \quad \text{for } i = 1,\ldots,n.
\end{align*}
$$
